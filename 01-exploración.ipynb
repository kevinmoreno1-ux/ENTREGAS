```python
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Specify the path to your CSV file in Google Drive
# Replace 'MyDrive/your_folder/your_file.csv' with your actual path
file_path = '/content/drive/MyDrive/IA/train.csv'

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv(file_path)



# Mostrar las primeras filas del archivo
print("Primeras filas del dataset:")
print(data.head())

print("\nInformación general del dataset:")
print(data.info())

print("\nColumnas disponibles:")
print(data.columns)

# Ejemplo 1: Cantidad de estudiantes por clase
if 'class' in data.columns:
    print("\nCantidad de estudiantes por clase:")
    print(data['class'].value_counts())

# Ejemplo 2: Cantidad de estudiantes por localidad
if 'localidad' in data.columns or 'location' in data.columns:
    columna_localidad = 'localidad' if 'localidad' in data.columns else 'location'
    print(f"\nCantidad de estudiantes por {columna_localidad}:")
    print(data[columna_localidad].value_counts())

# Ejemplo 3: Cantidad de estudiantes por combinación Clase - Localidad
if 'class' in data.columns and (('localidad' in data.columns) or ('location' in data.columns)):
    columna_localidad = 'localidad' if 'localidad' in data.columns else 'location'
    print(f"\nCantidad de estudiantes por clase y {columna_localidad}:")
    print(data.groupby(['class', columna_localidad]).size())

# Estadísticas generales
print("\nResumen estadístico:")
print(data.describe(include='all'))
```
